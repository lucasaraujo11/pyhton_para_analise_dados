```python
import pandas as pd
```

```python
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = f'https://raw.githubusercontent.com/lucasaraujo11/pyhton_para_analise_dados/refs/heads/main/datasets/babynames/yob{year}.txt'
    frame = pd.read_csv(path, names=columns)

    frame['year'] = year
    pieces.append(frame)

# Concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index=True)
```

```python
names.head()
```

```python
# top 10 nomes femininos mais comuns nos USA entre 1880 e 2011
names.query(' sex == "F" ').groupby(['name'])['births'].sum().nlargest(10).reset_index()
```

```python
names.query(' sex == "F" & year == 2010 ').groupby(['name'])['births'].sum().nlargest(10).reset_index()
```

```python
# top 10 nomes femininos menos comuns nos USA entre 1880 e 2011
names.query(' sex == "F" ').groupby(['name'])['births'].sum().nsmallest(10).reset_index()
```

```python
# top 10 nomes masculinos mais comuns nos USA entre 1880 e 2011
names.query(' sex == "M" ').groupby(['name'])['births'].sum().nlargest(20).reset_index()
```

```python
# top 10 nomes masculinos menos comuns nos USA entre 1880 e 2011
names.query(' sex == "M" ').groupby(['name'])['births'].sum().nsmallest(10).reset_index()
```

```python
names.query(' sex == "M" & year == 2010 ').groupby(['name'])['births'].sum().nlargest(10).reset_index()
```
